```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
my_orginal_img =cv2.imread("./binary.png",0)
```

```python
my_kernel = np.ones ((5,5),dtype = np.uint8)
my_kernel
```

```python
my_erosion_img = cv2.erode(my_orginal_img, my_kernel,iterations=4)

fig=plt.figure(figsize=(10,7))
rows=1
columns=2
fig.add_subplot(rows,columns,1)
plt.imshow(my_orginal_img,cmap="gray")
plt.axis('off')
plt.title("orginal image")

fig.add_subplot(rows,columns,2)
plt.imshow(my_erosion_img,cmap="gray")
plt.axis('off')
plt.title("Erosion image")
```

```python
my_dilation_img = cv2.dilate(my_orginal_img, my_kernel,iterations=2)

fig=plt.figure(figsize=(10,7))
rows=1
columns=3
fig.add_subplot(rows,columns,1)
plt.imshow(my_orginal_img,cmap="gray")
plt.axis('off')
plt.title("orginal image")

fig.add_subplot(rows,columns,2)
plt.imshow(my_erosion_img,cmap="gray")
plt.axis('off')
plt.title("Erosion image")

fig.add_subplot(rows,columns,3)
plt.imshow(my_dilation_img,cmap="gray")
plt.axis('off')
plt.title("dilation image")
```

```python
new_img = cv2.imread("./binary_with_text.png",0)

white_noise = np.random.randint(0,2,size=new_img.shape[:2])
white_noise = white_noise*255

```

```python
plt.imshow(white_noise,cmap="gray")
```

```python
my_noised_img=new_img + white_noise
plt.imshow(my_noised_img,cmap="gray")
```

```python
my_opening_img= cv2.morphologyEx(my_noised_img.astype(np.float32),cv2.MORPH_OPEN,my_kernel)
```

```python


fig=plt.figure(figsize=(10,7))
rows=1
columns=3

fig.add_subplot(rows,columns,1)
plt.imshow(new_img,cmap="gray")
plt.axis('off')
plt.title("orginal image")

fig.add_subplot(rows,columns,2)
plt.imshow(my_noised_img,cmap="gray")
plt.axis('off')
plt.title("My Noised image")

fig.add_subplot(rows,columns,3)
plt.imshow(my_opening_img,cmap="gray")
plt.axis('off')
plt.title("My Opening Image")
```
